```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
```

```python
def give(obj):
    li=[]
    for i in ast.literal_eval(obj):
      li.append(i['name'])
    return li
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres']=movies['genres'].apply(give)
```

```python
movies['keywords']=movies['keywords'].apply(give)
```

```python
def direct(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies.head(1)
```

```python
movies['cast']=movies['cast'].apply(direct)
```

```python
def get_director(obj):
  li=[]
  for i in ast.literal_eval(obj):
    if i["department"]=="Directing":
      li.append(i['name'])
      break
  return li
#"department": "Directing"
```

```python
movies['crew']=movies['crew'].apply(get_director)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=7000,stop_words='english')
```

```python
new_movies.head(1)
```

```python
cv.fit_transform(new_movies['tags']).toarray().shape
```

```python
vectors=cv.fit_transform(new_movies['tags']).toarray()
```

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def stem(text):
  li=[]
  for i in text.split():
    li.append(lemmatizer.lemmatize(i))
  return " ".join(li)
```

```python
new_movies['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
similarity[0]
```

```python
def recommend(movie):
  index = new_movies[new_movies['title'] == movie].index[0]
  distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
  for i in distances[1:6]:                                           # sorting on the basis of 2nd number
      print(new_movies.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
